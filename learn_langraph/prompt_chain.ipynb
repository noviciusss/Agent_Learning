```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_groq import ChatGroq
from dotenv import load_dotenv
load_dotenv()
```

```python
model = ChatGroq(model="llama-3.3-70b-versatile", temperature=0)
```

```python
class BlogState(TypedDict):
    title: str
    outline:str
    content:str
```

```python
def outline(state: BlogState)-> BlogState:
    title = state['title']
    prompt = f"""You are a world class content creator. Create a detailed outline for a blog post with the title: {title}.
    The outline should have at least 5 sections with 2-3 bullet points for each section."""
    
    response = model.invoke(prompt)
    state['outline'] = response.content
    return state
```

```python
def content(state: BlogState) -> BlogState:
    outline = state['outline']
    prompt = f"""You are a world class content creator. Write a detailed blog post based on the following outline:
    {outline}
    The blog post should be engaging and informative."""
    
    response = model.invoke(prompt)
    state['content'] = response.content
    return state
```

```python
graph = StateGraph(BlogState)

graph.add_node('Create_outline',outline)
graph.add_node('conetent',content)

graph.add_edge(START,'Create_outline')
graph.add_edge('Create_outline','conetent')
graph.add_edge('conetent',END)

workflowgraph= graph.compile()
```

```python
intial_state = {'title': 'The Future of Artificial Intelligence in Everyday Life'}
final_state = workflowgraph.invoke(intial_state)
print(final_state['outline'])
print("-----________________________________-----")
print(final_state['content'])
```
